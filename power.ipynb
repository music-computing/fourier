## Background

As we've seen, periodic signals can be expressed simply in terms of their period.
Noise is more complex as it is 'spread' across frequency range
(the full range in the case of 'white noise').

## Parseval's Power

Parseval describes the power of a vector
$X = (x_0, x_1, \dots, x_n)$ as
$$
|X|^2 = \sum_{j=0}^{n-1} x_j^2
$$

Parseval's theorem states that the power of a set and its DFT are equal up to a constant factor of $\frac{1}{n}$:

$$
|X|^2 = \frac{1}{n}|\hat{X}|^2
$$

```python
def parseval(arr: np.ndarray) -> float:
    """
    Implements Parseval's theorem, measuring the overall 'power' of an input vector.
    According to this theory, the power of a set and DFT thereof are equal up to a constant factor of 1/n.
    """
    return np.sum(np.square(arr))
```

```python
def demo_parseval_equivalence(signal: np.ndarray) -> bool:
    """
    Demonstrates the equivalence between time-domain and frequency-domain power.
    """
    time_power = parseval(signal)
    dft = np.fft.fft(signal)
    freq_power = np.sum(np.square(np.abs(dft))) / n

    return np.allclose(time_power, freq_power)
```

```python
# Generate a test signal which is the sum of two sines with different frequencies
n = 256
t = np.arange(n)
freq1, amp1 = 5, 1.0    # 5 cycles in N points
freq2, amp2 = 12, 0.5   # 12 cycles in N points

signal = amp1 * np.sin(2 * np.pi * freq1 * t / n) + \
         amp2 * np.sin(2 * np.pi * freq2 * t / n)

demo_parseval_equivalence(signal)
```
