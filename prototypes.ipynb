## Background

Harmonic qualities can be characterised by finding _prototypes_ for each coefficient.
These are sets that maximise a given coefficient within a given cardinality.
A set with a large value on a given coefficient will resemble the prototypes for that coefficient.

## Solutions

```python
def find_prototype(cardinality, k) -> list:
    """
    Find the prototype/s that maximise the magnitude of Fourier coefficient k
    among all pc sets of the given cardinality.

    Parameters:
    - `cardinality`: the number of pitch classes (2-11)
    - `k`: the Fourier coefficient index (1-11)

    Returns: A list of one or more pitch class set/s.
    """
    results = []
    best = 0
    for pc_set_info in pc_set_functions.set_classes_from_cardinality(cardinality):
        vector = multiset_to_vector(pc_set_info[1], max_index=12)
        dft = np.fft.fft(vector)
        mag = np.abs(dft[k])
        if mag > best:
            best = mag
            results = [pc_set_info[1]]
        elif mag == best:
            results.append(pc_set_info[1])

    return results
```

```python
# 1.
find_prototype(cardinality=3, k=1)
```

```python
# 2.
print(find_prototype(cardinality=2, k=2), "\n", find_prototype(cardinality=4, k=2))
```

```python
# 3.
print(find_prototype(cardinality=3, k=3), "\n", find_prototype(cardinality=6, k=3))
```

```python
# 4.
print(find_prototype(cardinality=4, k=4), "\n", find_prototype(cardinality=8, k=4))
```

```python
# 5.
cards = [5, 6, 7]
for c in cards:
    print(find_prototype(cardinality=c, k=5))
```

```python
# 6.
find_prototype(cardinality=6, k=6)
```
