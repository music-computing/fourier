## Background

The Fourier equation can be expressed in a number of ways.

Perhaps the most 'standard' expression is as follows:

$$
\hat{x}_k = \sum_{j=0}^{n-1} x_j e^{-i 2 \pi k j / n}
$$

It can also be written compactly in matrix form as $F = Af$, where $f$ is a time signal, $F$ is the discrete Fourier transform (DFT), and $A$ is an $N \times N$ matrix with coefficients

$$
a_{i,j} = \frac{1}{\sqrt{N}} e^{- i \frac{2\pi}{N} ij}
$$

Note that this matrix form uses a $\frac{1}{\sqrt{N}}$ normalisation, making $A$ a **unitary** matrix ($A A^* = I$). This differs from the sum formula above by that same factor; the two are otherwise equivalent.

For example, where $N=4$:
$$
    {\bf A} = \frac{1}{\sqrt{4}} \begin{pmatrix}
    1 & 1 & 1 & 1\\
    1 & -i & -1 & i\\
    1 & -1 & 1 & -1\\
    1 & i & -1 & -i
    \end{pmatrix}
$$

## Reference

### Task 1 — build the DFT matrix $A$

```python
def square_dft_matrix(n: int) -> np.ndarray:
    """
    Build the unitary DFT matrix A of size n x n.

    Multiplying A by a signal vector of length n gives the
    normalised DFT (scaled by 1/sqrt(n) relative to numpy.fft.fft).

    :param n: size of the square matrix.
    :return: DFT matrix of shape (n, n), dtype complex128.
    """
    A = np.zeros((n, n), dtype=np.complex128)
    for i in range(n):
        for j in range(n):
            A[i, j] = np.exp(-2j * np.pi * i * j / n)
    return A / np.sqrt(n)
```

### Task 2 — apply the matrix row-by-row (explicit loop over signal)

```python
def dft_dot_product_steps(x: np.array) -> np.ndarray:
    """
    Compute the normalised DFT of signal x by explicitly building
    each row of the DFT matrix and dotting it with x.

    This makes the matrix-vector multiplication transparent:
    for each output index i, we multiply x[j] by the corresponding
    matrix entry and sum.

    :param x: input signal, array of length n.
    :return: normalised DFT of x, array of length n, dtype complex128.
    """
    x = x.astype(np.complex128)
    n = len(x)
    result = []
    for i in range(n):
        total = 0
        for j in range(n):
            sq = np.exp(-2j * np.pi * i * j / n)
            total += x[j] * sq # NB: x[j] indexes the signal explicitly
        result.append(total)
    return np.array(result, dtype=np.complex128) / np.sqrt(n)
```

### Task 3 — combined function using `np.dot`

```python
def dft_at_once(x: np.array) -> np.ndarray:
    """
    Compute the normalised DFT of signal x using the DFT matrix.

    Builds the unitary DFT matrix A internally via `square_dft_matrix()`,
    then returns A @ x.

    :param x: input signal, array of length n.
    :return: normalised DFT of x, array of length n, dtype complex128.
    """
    x = x.astype(np.complex128)
    n = len(x)
    dft_mat = square_dft_matrix(n)
    return np.dot(dft_mat, x)
```

Verify that both implementations agree with each other and with NumPy (accounting for the $1/\sqrt{n}$ normalisation).

```python
a = np.array([1, 2, 3, 4])
np.allclose(
    dft_dot_product_steps(a),
    dft_at_once(a)
)
```
