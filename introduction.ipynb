## Background: The Discrete Fourier Transform

Suppose we have a pitch-class or beat-class set (*pc/bc-set*) with characteristic function

$$X = (x_0, x_1, x_2, \ldots , x_{n-1})$$

For a pitch-class set, $n = 12$; for a beat-class set $n$ might be 8, 12, 16, or some other integer.

The DFT consists of $n$ coefficients $\hat{x}_0, \hat{x}_1, \hat{x}_2, \ldots, \hat{x}_{n-1}$, defined as

$$
\hat{x}_k = \sum_{j=0}^{n-1} x_j \, e^{-i 2 \pi k j / n}
$$

Expanding via Euler's formula $e^{-i\theta} = \cos\theta - i\sin\theta$:

$$
\hat{x}_k = \sum_{j=0}^{n-1} x_j\bigl(\cos(2\pi kj/n) - i\sin(2\pi kj/n)\bigr)
$$

The resulting $\hat{x}_k$ is a complex number. Its real and imaginary parts are computed separately:

$$
\operatorname{Re}(\hat{x}_k) = \sum_{j=0}^{n-1} x_j \cos(2\pi kj/n)
$$

$$
\operatorname{Im}(\hat{x}_k) = -\sum_{j=0}^{n-1} x_j \sin(2\pi kj/n)
$$

## Reference Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dft(x):
    """
    Compute the Discrete Fourier Transform of sequence x,
    computing each coefficient by iterating over all items.

    Parameters
    ----------
    x : array-like of length n
        Input signal (real or complex).

    Returns
    -------
    X_hat : np.ndarray of shape (n,), dtype complex
        DFT coefficients x_hat_0 ... x_hat_{n-1}.
    """
    x = np.asarray(x, dtype=float)
    n = len(x)
    X_hat = np.zeros(n, dtype=complex)

    for k in range(n):
        real_part = 0.0
        imag_part = 0.0
        for j in range(n):
            angle = 2 * np.pi * k * j / n
            real_part +=  x[j] * np.cos(angle)
            imag_part += -x[j] * np.sin(angle)   # note the minus sign
        X_hat[k] = complex(real_part, imag_part)

    return X_hat
```

Verify that our local solution agrees with numpy.

Use a musical example of the pitch-class set for a C major triad `{0, 4, 7}`.

```python
pc_set = np.zeros(12)
pc_set[[0, 4, 7]] = 1

local_result  = dft(pc_set)
numpy_result = np.fft.fft(pc_set)

np.allclose(local_result, numpy_result)
```

Bonus: apply this to various pitch class sets and compare the results.

The magnitude $|\hat{x}_k|$ measures how evenly spaced the set.

Coefficient $k = 0$ is always the cardinality of the set.

```python
sets = {
    "C major triad {0,4,7}":        [0, 4, 7],
    "Diminished 7th {0,3,6,9}":     [0, 3, 6, 9],
    "Whole-tone scale {0,2,4,6,8,10}": [0, 2, 4, 6, 8, 10],
    "Chromatic cluster {0,1,2,3}":  [0, 1, 2, 3],
}

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
fig.suptitle("DFT magnitudes of pitch-class sets (n = 12)", fontsize=14, fontweight='bold')

for ax, (name, pcs) in zip(axes.flat, sets.items()):
    x = np.zeros(12)
    x[pcs] = 1
    mags = np.abs(dft(x))
    ax.bar(range(12), mags, color='steelblue', edgecolor='white')
    ax.set_title(name, fontsize=10)
    ax.set_xlabel("k")
    ax.set_ylabel(r"$|\hat{x}_k|$")
    ax.set_xticks(range(12))

plt.tight_layout()
plt.show()
```
